# **what is mean by Web Application Firewall ?**

**In AWS, WAF stands for Web Application Firewall. It is a security tool that helps protect your web applications from common threats and vulnerabilities like:**

**SQL injection** (where attackers try to manipulate your database),

**Cross-site scripting** (XSS) (where malicious scripts are injected into your site), and Other types of attacks that could exploit your web applications.

AWS WAF allows you to set rules to monitor and control incoming traffic to your applications, ensuring they stay secure while also allowing legitimate users to access them. It's like a security guard for your web apps!

**Other Common Web Attacks**

a. **Cross-Site Request Forgery (CSRF**):

What is it? Tricks authenticated users into performing unwanted actions on behalf of the attacker.

Example: A malicious link triggers a fund transfer or account update when clicked by an authenticated user.

Prevention: Use CSRF tokens, enforce SameSite cookies, and validate user intentions.

b. **Distributed Denial of Service (DDoS):**

What is it? Overwhelms a web application or service with an excessive amount of traffic to disrupt normal operations.

Example: A botnet floods a website, causing downtime.

Prevention: Use AWS Shield, AWS WAF rate-limiting rules, and scalable architecture.

c. **Broken Authentication:**

What is it? Weak authentication mechanisms allow attackers to hijack user accounts.

Example: Insecure password storage, predictable session IDs.

Prevention: Implement strong password policies, MFA, and secure session handling.

d. **Insecure Direct Object References (IDOR):**

What is it? Exposes sensitive resources by directly referencing them in URLs or parameters.

Example: http://example.com/account/123 where changing 123 gives access to another userâ€™s account.

Prevention: Use access controls and avoid exposing internal resource identifiers.

e. **Security Misconfiguration:**

What is it? Poorly configured servers or applications expose vulnerabilities.

Example: Default credentials, unpatched software, verbose error messages.

Prevention: Regular updates, strong configurations, and minimized permissions.

# **what is mean by Amazon CloudFront?**

**Amazon CloudFront is a Content Delivery Network (CDN) service in AWS. It helps deliver your website's content (like images, videos, or web pages) to users quickly and securely, no matter where they are in the world.**

**How it works:**

**Global Network of Edge Locations:** CloudFront has servers (called edge locations) around the world. When a user requests your content, CloudFront delivers it from the edge location closest to them, reducing latency and improving performance.

**Caching:** Frequently accessed content is cached at edge locations to speed up delivery and reduce the load on your origin server.

**Security:** It integrates with services like AWS WAF (Web Application Firewall) and supports HTTPS to protect your data.

**Example:**

If your website is hosted in a single AWS region (e.g., the US), users in other regions (e.g., Asia) might experience slow loading times. CloudFront caches your content globally, so users in Asia get the data from the nearest server, making your website faster for everyone.

In short, CloudFront makes your applications faster, more reliable, and secure for users worldwide!

#**what is mean by Amazon CloudTrail?**

**Amazon CloudTrail is a service in AWS that helps you track and log all activities (API calls) happening in your AWS account. It records who did what, when, and from where, providing full visibility into your account's actions.**

**Key Features:**

**Logs every API call** (e.g., creating a resource, deleting an instance).

**Tracks activities** done through the AWS Management Console, CLI, SDKs, or other AWS services.

**Stores logs** securely in Amazon S3 for analysis or auditing.

**Integrates** with Amazon CloudWatch for real-time monitoring and alerting.

**Example Use Case:**

Suppose a developer accidentally deletes an EC2 instance.

**You can use CloudTrail logs to:**

Identify who deleted the instance (username or role).

See when the deletion occurred.

Check from where the request originated (IP address or region).

#**Types of Events in CloudTrail**

**Management Events**

Track operations that manage AWS resources, such as creating, updating, or deleting resources.

**Example:**

Creating an EC2 instance (RunInstances).

Modifying IAM permissions (UpdateRole).

Deleting an S3 bucket (DeleteBucket).

**Data Events**

Record specific data-level operations for certain services.

Example:

Object-level activity in Amazon S3 (e.g., GetObject, PutObject).
Lambda function invocations.

These are not logged by default but can be enabled for more granular tracking.

**Insights Events**

Provide information about unusual or unexpected activity in your account.

Example: A sudden spike in API calls or unusual access patterns.

# **what is mean by AWS Trusted Advisor ?**

**AWS Trusted Advisor is a tool that checks your AWS account and gives you helpful tips to:**

**Save money**: It tells you if you're paying for things you don't need.

**Fix security problems:** It warns you if something is set up in a way that could be dangerous (like an open file).

**Improve performance:**It suggests ways to make your AWS services run faster.

**Avoid problems:** It helps you find things that could cause issues later, like running out of resources.

For example, if you have an unused server, Trusted Advisor might tell you to turn it off so you're not wasting money.
