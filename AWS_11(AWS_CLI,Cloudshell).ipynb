# **What is AWS CLI?**

AWS CLI (Command Line Interface) is a tool that lets you interact with AWS services using commands in a terminal or command prompt. It provides an easy way to manage AWS resources without using the AWS Management Console.

**Key Features:**

Manage AWS services like EC2, S3, Lambda, etc., from the command line.
Automate tasks using scripts.
Access and control AWS resources faster.


**Why Use AWS CLI?**

Efficiency: Perform tasks quickly without clicking through the console.
Automation: Combine commands into scripts to automate repetitive tasks.
Powerful: Access advanced AWS features directly.


**Steps to Generate Access and Secret Access Keys:**

**Log in to AWS Management Console:**

Go to AWS Console.

**Navigate to IAM (Identity and Access Management):**

In the search bar, type IAM and open it.

**Choose a User:**

Select the user for whom you want to generate keys.
If you don’t have a user, create one with programmatic access.

**Generate Keys:**

Go to the Security Credentials tab.
Click Create Access Key under the "Access Keys" section.

**Download Keys:**

Save the Access Key ID and Secret Access Key securely.
You won’t be able to see the Secret Access Key again.

**Configure AWS CLI with Keys:**

Use the following command to configure AWS CLI:

aws configure

Enter the Access Key ID, Secret Access Key, and default region when prompted.


**NOTE:**

The Access Key ID and Secret Access Key are generated only once during creation.

Make sure to copy both keys immediately after generating them.

If you lose the Secret Access Key, you cannot retrieve it again.

To regain access, you'll need to delete the old key and create a new one.

# **CloudShell: An AWS CLI Alternative**

AWS CloudShell is a browser-based shell provided by AWS to run CLI commands without needing to install or configure the CLI on your local machine.



**Why Use It?**

No need to install or set up AWS CLI on your computer.

Automatically uses your AWS account permissions.

Includes 1 GB of storage for saving files or scripts.
