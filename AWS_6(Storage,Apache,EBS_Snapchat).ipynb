**AWS Storage Types:**

**gp2 (General Purpose SSD):**
        
Balanced price & performance.
        
Good for boot volumes and small databases.

**io1/io2 (Provisioned IOPS SSDs):**
        
High-performance storage.
        
Use for large, critical databases (e.g., MySQL, SAP).

**SSD vs. HDD:**
        
SSD: Fast, good for apps and databases.
        
HDD: Slower, good for backups and large files.

**Storage Sizes:**

GiB (Gibibyte): Smaller unit (1 GiB = ~1.07 GB).
    
TiB (Tebibyte): Larger unit (1 TiB = ~1.1 TB).

**Example:**

Website storage: Use gp2 (fast and affordable).

Database: Use io2 (high performance).

Backups: Use HDD (cheap for large storage).

This setup balances cost and performance!

# **What is Apache and Its Use in AWS?**

Apache is a popular open-source web server software that helps deliver web content to users over the internet. In AWS, it is commonly used with EC2 instances to host websites and web applications.

# **Uses of Apache in AWS:**

**Host Websites:**
        
Apache serves static and dynamic web pages (HTML, PHP, etc.) to users.

**Run Web Applications:**
        
Supports frameworks like PHP, Python, or Ruby for backend web apps.

**Reverse Proxy:**
        
Can act as a gateway between users and backend servers, improving scalability.

**Load Balancing:**
        
Distributes traffic across multiple servers to improve performance and reliability.

**Custom Configuration:**
        
Allows setting up virtual hosts, SSL/TLS encryption, URL rewriting, and more.

**Example:**

    You launch an EC2 instance in AWS to host a blog.
    Install Apache to serve the blog content to visitors.
    Apache handles requests like displaying blog pages or processing form submissions.

Summary:

Apache is used in AWS for hosting websites, running applications, managing traffic, and securing web content. It's flexible, reliable, and widely supported.

# **How to Set Up Apache on AWS Linux (Step-by-Step)**

**Launch an EC2 Instance:**
        
Go to AWS Console, create an EC2 instance (e.g., Amazon Linux or Ubuntu).
        
Attach a security group with ports 22 (SSH) and 80 (HTTP) open.

**Connect to the Instance:**
        
Use SSH to connect:

    ssh -i your-key.pem ec2-user@your-ec2-public-ip

# **Install Apache:**

**Update packages:**

sudo yum update -y      # For Amazon Linux

sudo apt update         # For Ubuntu

**Install Apache:**

sudo yum install httpd -y  # Amazon Linux

sudo apt install apache2   # Ubuntu

**Start Apache Service:**

    Start Apache and enable it to run at boot:

sudo systemctl start httpd       # Amazon Linux

sudo systemctl enable httpd

**Or for Ubuntu:**

sudo systemctl start apache2

sudo systemctl enable apache2

**Verify Apache Installation:**

    Open your browser and go to the EC2 public IP (e.g., http://your-ec2-public-ip).
    You should see the Apache welcome page.

**Set Up a Web Page (Optional):**

    Replace the default Apache page with your content:

echo "<h1>Welcome to My Website</h1>" | sudo tee /var/www/html/index.html

**Reload Apache to apply changes:**

sudo systemctl restart httpd    # Amazon Linux
        
sudo systemctl restart apache2 # Ubuntu

**Adjust Security Group if Needed:**
        
        Ensure your instance's security group allows HTTP (port 80) traffic.

Summary:

    Launch EC2 and SSH in.
    Install Apache.
    Start Apache and verify via browser.
    Optionally, customize your web page.

This is how you host a basic website on AWS using Apache!



# **What are EBS Snapshots?**

EBS Snapshots are backups of your EBS volumes.
    
They save the data on an EBS volume at a specific point in time.
    
Stored in Amazon S3 (managed by AWS), they are used for backup, recovery, and creating new EBS volumes.

# **What is mean by EBS Volume?**

**Amazon Elastic Block Store (EBS), a "volume" refers to a virtual hard disk or block-level storage device that you can attach to your Amazon EC2 (Elastic Compute Cloud) instances. These EBS volumes are used to store data persistently, meaning the data remains even after the EC2 instance is stopped or terminated (unless the volume is deleted).**

**In AWS, EBS snapshots are point-in-time backups of your EBS volumes. You can share these snapshots with others, but there are considerations and potential disadvantages to keep in mind.**

**Sharing EBS Snapshots**

**Private Sharing:** You can share snapshots with specific AWS accounts by specifying the account ID.
    
**Public Sharing:** You can make a snapshot public, allowing anyone to access and use it.



# **Why is it called a "point-in-time" snapshot?**

**It captures the state of your data at a specific instant, just like taking a photo freezes a moment.**

**Encrypted Snapshot:** A snapshot where data is scrambled using a key to protect it from unauthorized access. Example: A snapshot of a database with sensitive user passwords stored securely.

**Decrypted Snapshot:** A snapshot where encrypted data is converted back into readable form using the correct decryption key. Example: Restoring the above snapshot to access the original database for analysis.
