# **what is mean by server?**

**a server in AWS is like a computer that runs in the cloud to do a specific job. Instead of being a physical computer you own, AWS provides it to you virtually, and you only pay for what you use.**

**it is nothing but a virtual machine we can access the information through the internet.**

**Example:Imagine your company creates websites for other businesses. To deliver this service, you need a place to host these websites so they are accessible 24/7. Instead of buying physical servers, you use AWS servers.**

Example:

**Client Request:**
  
A client company asks you to build and host their e-commerce website where they can sell products online.

**AWS Solution:**
        
You log in to AWS and create an EC2 instance (a virtual server).
        
You install the necessary software (e.g., a web server like Apache or NGINX) on the EC2 instance.
        
You upload the client’s e-commerce website files to the server.

**Website Goes Live:**
        
The EC2 instance runs the website.
        
People can now access the e-commerce site by typing the domain (e.g., www.clientwebsite.com) into their browser.

**Selling to Other Companies:**
        
When another client comes to you, you repeat the process by creating a new EC2 instance for their website or hosting multiple client websites on the same server, depending on the size and traffic.

# **what is mean by EC2?**

**EC2 in AWS stands for Elastic Compute Cloud. It’s like a virtual computer in the cloud where you can run applications, websites, or software.**(It is mostly used to creating the servers).

**Simple Example:**

Imagine you need a computer to host a website or run a program, but instead of buying a physical one, AWS gives you a virtual computer.
    
You can choose how powerful it is (CPU, memory, storage), turn it on when you need it, and only pay for the time you use it.

EC2 makes it easy to get a server quickly, without worrying about hardware.

**Here’s a simple step-by-step process for creating a server to sell websites to other companies using AWS:**


**Set Up the Server**

Use AWS EC2 to create a virtual server.
    
Choose the server specifications (like operating system, CPU, and memory).

**Install the Website**

Upload your website files (HTML, CSS, images, or CMS like WordPress) to the server.
    
Test the website to make sure it works properly.

**Provide a Domain Name**

Use AWS Route 53 or another domain provider to link a domain name (e.g., www.clientwebsite.com) to the server.
    
This makes the website easier to access instead of using an IP address.

**Open the Server for Access**

Configure the security group for the EC2 instance to allow access through port 80 (HTTP) or 443 (HTTPS).
    
This ensures the website is publicly accessible.

**Share Access with Clients**

Provide the client with the website link (e.g., www.clientwebsite.com).
    
The client can now use the website or give it to their customers.

**Sell and Manage More Websites**

For each new client, repeat the process:
        
Create a new server or use the same server if resources allow.
        
Add their specific website.
        
Link their domain name.

# **what is mean by On-Demand Instance ?**

**An On-Demand Instance in AWS is a virtual server that you can rent and pay for only when you use it. There’s no upfront payment or long-term commitment—you just pay for the time the instance is running.**

EX:It’s like booking a hotel room for a night: you use it when you need it and stop paying when you’re done. This is great for short-term or unpredictable workloads.


# **what is mean by  Reserved Instance (RI) ?**

aws9.jpeg

**A Reserved Instance (RI) in AWS is a virtual server you can rent at a discounted price by committing to use it for a specific period, typically 1 year or 3 years.**

Key Points:

**Cheaper:** You get a lower price compared to On-Demand Instances because of the long-term commitment.
    
**Predictable Workloads:** Ideal for applications you know will run continuously or for a long time.

**Example:**

If you have a web application that will run 24/7 for the next 3 years, buying a Reserved Instance can save you money. For example, instead of paying $50/month for an On-Demand Instance, you might pay $35/month with a Reserved Instance.


# **what is Spot Instance?**

**A Spot Instance in AWS is a virtual server that lets you use unused AWS capacity at a much lower cost than On-Demand Instances.**

Key Points:

**Cheaper:** You can save up to 90% compared to On-Demand prices.
    
**Unpredictable:** AWS can take back the instance anytime if it needs the capacity, so it’s not suitable for critical workloads.
    
**Best for Flexible Workloads:** Good for tasks that can handle interruptions, like batch processing, big data analysis, or testing.

**Example:**

If you’re running a data analysis job that takes 6 hours and can restart if interrupted, you can use a Spot Instance to save money instead of paying full price for an On-Demand Instance.


note:A Spot Instance is a server in AWS that costs much less because it uses spare capacity. If the regular price is $50, you might get it for $15–$35. But AWS can take it back anytime if they need it, so it’s good for tasks that can handle interruptions.
