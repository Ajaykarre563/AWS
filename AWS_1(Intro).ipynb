# **what is mean by aws?**

**AWS, or Amazon Web Services, is a platform that provides a wide range of cloud computing services.**

In simple terms, it’s like renting powerful computers, storage, databases, and tools over the internet instead of buying and managing them yourself. This helps businesses and individuals build and run applications without needing physical hardware.

# **what is mean by cloud computing?**

**Cloud computing means using the internet to access services like storage, software, or servers instead of having them on your own computer or devices.**

It’s like storing your files in Google Drive or using Netflix—everything runs on someone else’s computers (the "cloud"), and you can access it anytime, anywhere with an internet connection.

# **AWS is widely used**

AWS is widely used across various industries and is especially useful in companies like:

**Tech Companies** - For hosting apps, websites, and managing data. Examples: Netflix, Airbnb.
    
**E-commerce** - To handle online stores, customer data, and payment systems. Example: Amazon (its own business), Shopify.
    
**Finance and Banking** - For secure data processing and analytics. Example: Capital One, Stripe.
    
**Healthcare** - For storing medical records and running healthcare applications. Example: Pfizer.
    
**Media and Entertainment** - For streaming content and storing large video files. Example: Disney+, Spotify.
    
**Startups** - To quickly and cheaply build and scale their applications.

AWS is popular because it offers flexibility, reliability, and cost savings for businesses of all sizes.


# **Advantages of AWS**

**Scalability:** Easily scale resources up or down based on your needs.
    
**Cost-Effective:** Pay-as-you-go pricing eliminates upfront hardware costs.
    
**Global Reach:** Data centers worldwide for low latency and better performance.
    
**Wide Range of Services:** Offers 200+ services, from storage to AI and machine learning.
    
**Reliability**: Highly reliable infrastructure with backups and failover options.
    
**Security:** Built-in security measures, compliance with industry standards, and encryption.
    
**Flexibility:** Works with many operating systems, programming languages, and tools.

# **Disadvantages of AWS**

**Complex Pricing:** The cost structure can be confusing, especially for beginners.
    
**Downtime Risks:** Rare, but outages can affect your applications.
    
**Data Transfer Costs:** Moving data out of AWS can become expensive.
    
**Learning Curve:** Some services are complex and require expertise to use effectively.
    
**Vendor Lock-In:** Once you're heavily invested, switching to another platform can be difficult.
    
**Support Costs:** Detailed customer support comes with extra charges.



# **What is mean by cloud and its types?**

**A cloud in computing refers to using internet-based servers to store data, run applications, and provide various IT services. Instead of relying on your personal computer or local servers, the "cloud" lets you access resources online from anywhere.**

**Types of Clouds:**

**Public Cloud**(everyone created account we can data centers)
        
Owned and operated by third-party providers like AWS, Microsoft Azure, or Google Cloud.
        
Accessible to anyone who wants to pay for services like storage, computing, or software.

**Private Cloud**(we can request to aws to create my own server they are used by only my organisation)
        
Exclusive to a single organization.
        
Used for enhanced security and control, typically hosted either on-premises or by a private cloud provider.

**Hybrid Cloud**(4 or 5 company are bulid they own server centers)
        
Combines both public and private clouds.
        
Allows businesses to use private clouds for sensitive data and public clouds for less critical tasks.

**Community Cloud**
        
Shared by multiple organizations with similar needs (e.g., government agencies).
        
Helps reduce costs while maintaining shared security and compliance standards.

Clouds provide flexibility, scalability, and cost-effectiveness for storing and running applications!

# **what is mean by cloud service models**

**Cloud service models describe how cloud computing resources and services are delivered and managed. They determine the level of responsibility shared between the user and the cloud provider. There are three main cloud service models: IaaS, PaaS, and SaaS.**

**On-Premises**

IT is refers to a traditional IT infrastructure setup where all hardware, software, and resources are physically located and managed at the organization's site (like in a company's data center). It is different from cloud computing, where resources are hosted and managed off-site by providers like AWS.

# **Infrastructure as a Service (IaaS)**

AWS offers virtualized computing resources like servers, storage, and networking.
    
You manage the operating system, apps, and data, while AWS handles the hardware.

**AWS IaaS Services:**

Amazon EC2: Virtual servers.
    
Amazon S3: Scalable storage.
    
Elastic Load Balancing (ELB): Distributes traffic.

**Advantages:**

High flexibility and control.

Pay only for the resources used.

**Use Cases:**

Hosting websites, data storage, and custom app development.

# **2. Platform as a Service (PaaS)**

AWS provides a platform for building, testing, and deploying applications.
    
AWS manages the underlying infrastructure, operating systems, and runtime.

**AWS PaaS Services:**

AWS Elastic Beanstalk: Automatically deploys and scales web applications.
    
AWS Lambda: Runs your code without provisioning or managing servers.

**Advantages:**

Simplifies development and deployment.
    
Focus on application development instead of managing infrastructure.

**Use Cases:**

Web and mobile application development.
    
Rapid app prototyping.

# **3. Software as a Service (SaaS)**

AWS hosts software applications that users can access over the internet.
    
AWS or third-party providers handle the infrastructure, platform, and software maintenance.

**AWS SaaS Examples:**

Amazon WorkSpaces: Virtual desktops.
    
Amazon Connect: Cloud contact center solution.

**Advantages:**

No need to install or maintain software.
    
Easy scalability and accessibility.

**Use Cases:**

Customer support systems, collaborative tools, and virtual workspaces.

# **History of AWS**

Here’s a brief history of AWS in one line for each key milestone:

2002: AWS launched with Simple Queue Service (SQS), its first service.
    
2004: Elastic Compute Cloud (EC2) was introduced, enabling on-demand compute capacity.
    
2006: AWS officially launched with S3 (Simple Storage Service) for scalable cloud storage.
    
2007: Amazon Elastic MapReduce (EMR) was introduced for big data processing.
    
2009: Amazon CloudFront (CDN) and Elastic Load Balancing services were launched.
    
2012: AWS expanded its services with AWS Lambda for serverless computing.
    
2014: AWS introduced the Amazon Aurora relational database engine.
    
2015: AWS IoT was launched to connect Internet of Things devices.
    
2016: AWS began its global expansion with new regions and availability zones.
    
2018: AWS Deep Learning AMIs and AI services like Lex and Polly were introduced.
    
2020: AWS hit a significant milestone of $35 billion in annual revenue.
