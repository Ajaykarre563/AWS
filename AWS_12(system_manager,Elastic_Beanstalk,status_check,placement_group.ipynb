# **what is mean by AWS Systems Manager?**

AWS Systems Manager is a service designed to help you manage and operate your infrastructure efficiently and securely. Here’s a breakdown of the system’s

**core functions:**

Automation: Systems Manager allows you to automate common tasks, such as software deployment, system updates, and maintenance across multiple machines or environments. This helps reduce manual errors and save time.

Centralized Management: With Systems Manager, you can manage both AWS resources (like EC2 instances) and on-premises servers from a central location. This is useful for large-scale environments where manual management would be impractical.

Security and Compliance: The service offers features to improve security, such as managing sensitive information securely with Parameter Store and enforcing patch compliance to keep your systems up-to-date and secure.

Efficient Operations: It provides a set of tools that streamline operational tasks:

Run Command: Remotely execute scripts or commands on your resources.
Patch Manager: Automatically manage software patches across your environment.
State Manager: Ensure systems are maintained in a specified configuration or state.

Operational Insights: You can use Systems Manager’s features like Inventory and Session Manager to collect data, troubleshoot, and establish secure connections to instances without needing direct network access.

# **what is mean by get log ?**

**"get log" refers to retrieving or accessing logs that record the output or details of executed commands, events, or activities. Logs provide insights into what happened during the execution of a command, helping with troubleshooting, auditing, or monitoring.**

# **How to Retrieve Logs in AWS Systems Manager**

**Command Output Logs:**

In the AWS Management Console:
Go to the Run Command section.
Select the command from the Command History.
View the output logs directly in the console for each target instance.

**CloudWatch Logs Integration:**

If you've configured the command to send output to Amazon CloudWatch Logs, you can:
Open the CloudWatch service.
Navigate to the appropriate log group and stream to view the logs.

**S3 Bucket Integration:**

If the output is stored in an S3 bucket:
Navigate to the S3 service.
Locate the specified bucket and folder to access the log files.

# **what is mean by AWS Elastic Beanstalk?**

**AWS Elastic Beanstalk is a service that makes it easy to host and manage your applications in the cloud without worrying about servers or infrastructure.**

**Example:**

You have a web app written in Python (Flask).

You zip your app files (code and dependencies).

You upload the zip file to Elastic Beanstalk.

Elastic Beanstalk automatically:

Creates servers.

Installs the software your app needs.

Sets up scaling (to handle more users if traffic increases).

It gives you a URL to access your app.

You focus on coding, and Elastic Beanstalk takes care of running your app online.

# **what is mean by Status Check in AWS ?**

**Status Check in AWS**

Status checks in AWS ensure that your EC2 instances are healthy and running properly. There are two types:

**System Status Check:**

Checks the health of the physical server hosting your instance.
If there’s an issue (e.g., network or hardware failure), AWS will try to fix it, or you can restart the instance.

**Instance Status Check:**

Checks the operating system and configuration inside your instance.
If there’s an issue (e.g., full disk or misconfigured network), you need to fix it yourself by logging into the instance.

# **what is mean by Placement Group ?**

**A Placement Group controls how your EC2 instances are arranged on AWS hardware to improve performance or reliability.**

There are three types:

**Cluster:**

Instances are placed very close together (same rack).
Best for apps needing fast communication and low delay (e.g., high-performance computing).



**Spread:**

Instances are placed on different racks.
Best for important workloads where you want to reduce the risk of failures.

**Partition:**

Instances are divided into isolated groups (partitions).
Best for big data apps like Hadoop where isolation between groups is needed.
