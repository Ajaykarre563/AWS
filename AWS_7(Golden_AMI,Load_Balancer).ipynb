# **What is a Golden AMI?**

A Golden AMI is a pre-configured template in AWS that includes everything you need to run your applications, such as:

The Operating System (e.g., Linux, Windows).
Required software (e.g., web servers like Apache).
Security settings and other configurations.
Its like a ready-made blueprint for creating identical servers.

**Real-Life Use**

For an e-commerce site:

A Golden AMI can include a database, app code, and monitoring tools.
Every new server launched will meet the companys standards without extra work.


# **What is a Load Balancer in AWS?**

A Load Balancer in AWS is a service that automatically distributes incoming traffic across multiple servers (EC2 instances). It ensures no single server is overloaded and improves the availability, reliability, and scalability of your application.

**Why Use a Load Balancer?**

**Traffic Distribution:** Balances the load between multiple servers to prevent overloading.

**High Availability:** If one server fails, traffic is sent to other healthy servers.

**Scalability:** Helps manage increasing traffic by adding more servers automatically.

**Simple Example**

Imagine you have a website hosted on 3 EC2 instances.
During high traffic (e.g., during a sale), one server might struggle.
A Load Balancer will divide the traffic among all 3 servers to ensure smooth performance.

If one server goes down, the Load Balancer sends traffic to the remaining healthy servers.

**Types of Load Balancers in AWS**

AWS provides three main types of load balancers under the Elastic Load Balancing (ELB) service:

**Application Load Balancer (ALB):**

Works at the HTTP/HTTPS (Layer 7) level.

Best for web applications.

Example: Routes requests to different servers based on URL paths (e.g., /login goes to one server, /checkout to another).

**Network Load Balancer (NLB):**

Works at the TCP/UDP (Layer 4) level.

Best for low-latency, high-performance use cases.
Example: Ideal for gaming servers or real-time systems.

**Gateway Load Balancer (GWLB):**

Works at Layer 3 (Network Layer).

Purpose: Routes network traffic through third-party virtual appliances (e.g.,
firewalls, intrusion detection systems, or monitoring tools).

Example:

Inspect incoming traffic for malicious activity using a firewall.
Gateway Load Balancer directs traffic to the firewall instances for inspection before allowing it to proceed to your application.

It is ideal for integrating security and traffic inspection tools into your AWS environment while ensuring scalability and high availability.

**Classic Load Balancer (CLB):**

Older version; supports both Layer 4 (TCP) and Layer 7 (HTTP/HTTPS).
Example: Used for basic load balancing tasks (less commonly used now).

**Key Benefits**

Prevents downtime.
Improves user experience.
Automatically scales to handle traffic spikes.

**Note:**

If any server fails, the Load Balancer does not send traffic to the failed (damaged) server. Instead, it redirects traffic only to the servers that are healthy and working properly.

**In AWS, a target group is a way to group resources like EC2 instances, containers, or IP addresses that handle incoming traffic. It is used with load balancers (e.g., Application Load Balancer or Network Load Balancer) to distribute traffic across these resources.**

**For example:**

If you have multiple servers (EC2 instances) running your application, you can group them in a target group.
    
The load balancer sends incoming requests to the target group, ensuring traffic is distributed efficiently.

## **A load balancer checks target health by periodically sending requests (e.g., HTTP, TCP) to a configured endpoint and routing traffic only to targets that respond as healthy.**


**If the Healthy Threshold is set to 2 and the target passes health checks 3 consecutive times, it will already be marked healthy after the second successful check. The extra check (the 3rd one) doesn't affect the status further.**

The key point is that the target needs to pass the specified threshold (2 in this case) to transition to the healthy state.
