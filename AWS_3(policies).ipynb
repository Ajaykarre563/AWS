# **what is mean by policies in AWS?**

**In AWS (Amazon Web Services), "policies" are rules or permissions that define what actions a user, group, or role is allowed or denied within the AWS environment These policies help control access to AWS resources securely.**

**Here are the types of AWS policies, each with a simple explanation and an example:**

**Identity-based Policies**
        
Attached to users, groups, or roles to define their permissions.
        
Example: Allowing a user to launch EC2 instances.

**Resource-based Policies**
        
Attached directly to resources to control access to them.
        
Example: An S3 bucket policy allowing public read access to files.

**Permissions Boundaries**
        
Limit the maximum permissions an identity can have.
        
Example: Preventing a role from deleting resources, even with other permissions.

**Service Control Policies (SCPs)**
        
Manage permissions across accounts in AWS Organizations.
        
Example: Denying all accounts the ability to disable CloudTrail.

**Session Policies**
        
Temporary permissions applied during a session.
        
Example: A developer assumes a role with permissions to access DynamoDB for one hour.

**Access Control Lists (ACLs)**
        
Provide simple permissions for specific resources.
        
Example: Allowing a specific IP address to access a VPC or an S3 object.



# **IAM Policies: These are rules that define what actions (like reading, writing, or deleting resources) a user, group, or role is allowed or denied to perform on specific AWS resources. Think of them as "permission rules" written in JSON.**

**Here's a simpler way to understand Denied Permission in AWS:**

**Allow means "You can do it."**
    
**Deny means "You cannot do it."**

If there is a rule that says "Allow" but another rule says "Deny," Deny always wins.

For example:

**One rule says, "You can delete the S3 bucket."**
    
**Another rule says, "You cannot delete the S3 bucket."**

Since "Deny" is stronger, you will not be allowed to delete the bucket.

# **what is mean by inline policy**

**An inline policy in AWS is a permission policy directly attached to a specific user, group, or role. It's like a set of rules that tells AWS what that identity can or cannot do.**

**Key Points:**

It belongs only to that specific identity.
    
It cannot be shared or reused with others.
    
If the identity is deleted, the policy is also deleted.
