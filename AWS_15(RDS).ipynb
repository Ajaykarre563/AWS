# **what is mean by RDS?**

**RDS in AWS stands for Relational Database Service. It’s a managed database service that helps you set up, operate, and scale relational databases in the cloud without needing to handle the underlying infrastructure.**

# Here’s the simple understanding:

Managed Service: AWS handles tasks like hardware setup, software installation, backups, patching, and scaling for you.

Relational Databases: It supports popular database engines like MySQL, PostgreSQL, SQL Server, Oracle, and Amazon Aurora.

Ease of Use: You can quickly create a database, connect your application to it, and focus on your app instead of managing the database backend.

Scalable: You can easily adjust the size and performance of the database as your needs grow.

# **Summary:**

**The diagram depicts a Multi-AZ (Availability Zone) Deployment for an Amazon RDS instance in AWS.**


Here’s a breakdown of the components:

1.**VPC (Virtual Private Cloud):**
The entire infrastructure is deployed within a VPC, which provides network isolation and security.

Inside the VPC, there are subnets and other resources organized for better
performance and availability.

2.**Subnets:**
Public Subnets (green boxes):
These are exposed to the internet and host the Jump Server.
The Jump Server allows secure access to resources in private subnets, often via SSH.

**Private Subnets (blue boxes):**

These are isolated from the internet and host the primary and standby RDS instances.

Private subnets improve security by ensuring database instances are not directly accessible from the internet.

3.**Multi-AZ Deployment:**
This ensures high availability and fault tolerance by replicating data between two RDS instances in different Availability Zones (AZs):

Primary RDS Instance (M):
Handles both read and write operations.

Secondary RDS Instance (S):
Acts as a standby instance, automatically promoted to the primary role if the primary instance fails.
The replication between primary and standby is managed by AWS.

4.**Jump Server:**
Acts as a secure access point for administrators or users to connect to resources within the private subnets.
Typically used to manage and monitor the infrastructure.

5.**Network Communication:**
The diagram shows IP addresses (e.g., 192.168.2.136 and 192.168.2.78) for the private subnets where the RDS instances reside.
Network routing ensures the two RDS instances in different AZs can communicate with each other securely and efficiently.

6.**Additional Components (Red Boxes at the Bottom):**
Likely represent firewalls or network gateways for securing access to the environment.

**Purpose of this Setup:**

This Multi-AZ setup is designed to ensure high availability, reliability, and fault tolerance for critical applications. If the primary instance fails, AWS automatically switches to the standby without manual intervention, minimizing downtime.

# **Types of Snapshots in AWS RDS:**

**Automated Snapshots:**

Created automatically by RDS during the backup window you define.
These are part of the automated backup feature and allow point-in-time recovery.
Retention period: You can configure how many days these backups are retained.

**Manual Snapshots:**

Created manually by the user whenever needed.
These snapshots are retained indefinitely until you delete them.
Useful for long-term retention, data migration, or before making significant changes to the database.
