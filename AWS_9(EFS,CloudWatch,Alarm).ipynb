**What is Elastic File System (EFS) in AWS?**

Amazon Elastic File System (EFS) is a fully managed, scalable, and shared file storage service in AWS. It provides a file system that multiple EC2 instances can access at the same time, similar to a shared network drive.

**Key Features:**

Scalable: Automatically grows or shrinks based on the amount of data stored.

Shared Access: Multiple EC2 instances can read and write to the same file system simultaneously.

Managed Service: AWS handles maintenance, backups, and scaling.

High Availability: Data is stored across multiple Availability Zones for durability.

**Uses of Elastic File System (EFS):**

**Web Hosting:**

Store and serve website files that multiple servers can access.
Data Sharing:

Allow multiple EC2 instances to share files (e.g., logs, configurations).

**Big Data Analytics:**

Store and process large datasets that need to be accessed by many systems.

**Application Development:**

Host shared libraries or configurations used by multiple applications.

**Backup and Recovery:**

Use EFS as a durable and scalable backup solution.

**Simple Example:**

Imagine you have a photo-sharing website with multiple servers handling user uploads.

EFS allows all servers to access the same storage for uploaded photos, ensuring consistent access across the system.


**What is Amazon CloudWatch?**

**Amazon CloudWatch is a tool in AWS used to monitor and manage your resources and applications. It tracks performance, collects logs, and alerts you if something needs attention.**

**What is CloudWatch Used for in Real-Time?**

**Monitoring Resources:**

Checks the health and performance of servers, databases, and storage.
Setting Alerts:

Sends notifications (like email or SMS) if something goes wrong, like high CPU usage.

**Log Tracking:**

Collects application logs to help you find and fix issues.
Scaling Applications:

Helps automatically add more servers when traffic increases.

**Dashboards:**

Shows real-time data in graphs for easy understanding.

**Example in Real-Time:**

**Scenario:**

You run a shopping website.
CloudWatch monitors the servers.
If CPU usage exceeds 80%, it sends an alert.
It can also automatically add more servers to handle traffic.

**How to Create a Dashboard in Amazon CloudWatch**

A CloudWatch Dashboard is a customizable visual display of your application's metrics, making it easy to monitor performance in one place. You can also share dashboards with your team.

**Steps to Create a CloudWatch Dashboard:**

**Go to CloudWatch Console:**

Log in to your AWS Management Console.

Navigate to CloudWatch.

**Create a New Dashboard:**

Click on Dashboards in the left menu.

Click Create Dashboard and give it a name.

**Add Widgets:**

Select the type of widget you want to display (e.g., line graph, number, text).
Choose the metrics you want to monitor (e.g., CPU usage, memory, network traffic).
Customize the widget (e.g., time range, dimensions).

**Save the Dashboard:**

Arrange the widgets as needed.
Click Save Dashboard to keep your settings.

**How to Share a CloudWatch Dashboard:**

Public Sharing:

Dashboards can be made publicly accessible, but this requires enabling permissions carefully.

Internal Sharing:

Share it with team members by granting permissions using IAM roles or user groups in your AWS account.


**Example Use Case:**

You create a dashboard showing:

CPU usage of your EC2 instances.

Request count on your web servers.

Error logs from your application.
Share the dashboard with your DevOps team so everyone can monitor it in real-time.

**Note:**

Through a URL, we can access the CloudWatch Dashboard. AWS provides a shareable link for the dashboard, allowing team members with the necessary permissions to view it easily. This is useful for collaboration and real-time monitoring across teams.

**What is an Alarm in AWS?**

An Alarm in AWS is a feature of Amazon CloudWatch that tracks specific metrics and notifies you or takes automated actions when those metrics meet certain conditions.

**How Does It Work?**

You define a threshold for a metric (e.g., CPU usage > 70%).
The alarm continuously monitors the metric.

If the threshold is crossed, it triggers an action, such as:
Sending a notification (email, SMS, etc.).

Starting/stopping an instance.

Triggering Auto Scaling.

**Example:**

**Scenario:**

You have a web application running on an EC2 instance.

**What You Monitor:**

CPU usage.

Alarm Setup:

If CPU usage exceeds 80%, send an alert to your email.

If CPU usage stays high, trigger Auto Scaling to add more servers.

**Result:**

You’re notified of potential performance issues and can automatically handle increased traffic.

**Benefits:**

Helps detect problems early.

Automates responses to maintain application performance.

Reduces manual monitoring efforts.
